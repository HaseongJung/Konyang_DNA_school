```python
import cv2 #Computer Vision의 약자 버전 2
import numpy as np
import matplotlib.pyplot as plt

musk = cv2.imread('../data/elonMusk.jpg')
plt.imshow(musk)
plt.show()

musk = np.array(musk)
print(musk.shape)
```

```python
red_musk = musk[:,:,0]
green_musk = musk[:,:,1]
blue_musk = musk[:,:,2]

plt.imshow(red_musk,cmap='Reds'),
plt.title('R')
plt.show()

plt.imshow(green_musk,cmap='Greens'),
plt.title('G')
plt.show()

plt.imshow(blue_musk,cmap='Blues'),
plt.title('B')
plt.show()
```
